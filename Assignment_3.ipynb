```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants

def Speed(v, theta):
    c = constants.c
    v_parallel = (0.99 * c) + v
    v_perpendicular = v * np.sin(theta)
    v_total = np.sqrt(v_parallel**2 + v_perpendicular**2)
    return v_total / (1 + (v_total / c)**2 * 0.99 * c)


Mary_Speed = np.linspace(0, 0.99 * constants.c, 100)


theta = np.radians(30)

proton_speed = Speed(Mary_Speed, theta)


fig, ax = plt.subplots(figsize=(7, 7))
ax.plot(Mary_Speed, proton_speed)
ax.set_xlabel("Mary speed (m/s)")
ax.set_ylabel("Proton speed (m/s)")
plt.title("Proton Speed with angles")
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

c = 3*10**5

flash1_location = 0
flash2_location = 300000
jwst_location = 150000

def aliens(t):
  return 500000-0.1*c*t

t = np.linspace(0, 2 * flash2_location / c, 100)

plt.figure(figsize=(8, 6))

plt.plot(flash1_location, c * 0, 'ro', label='Flashbulb 1 (Earth)')
plt.plot(flash2_location, c * 0, 'bo', label='Flashbulb 2')


x_range = np.linspace(-flash2_location, 2 * flash2_location, 200)

plt.plot(x_range + flash1_location, c * np.abs(x_range), 'r--', label='Light from Flashbulb 1')
plt.plot(x_range + flash2_location, c * np.abs(x_range), 'b--', label='Light from Flashbulb 2')
plt.plot(aliens(t), t*0,label="Microprobe", color='purple')

plt.axvline(x=jwst_location, color='green', linestyle='-', label='JWST')

plt.xlabel('x (km)')
plt.ylabel('ct')
plt.title('Spacetime Diagram with Light from Flashbulbs and JWST')
plt.legend()
plt.grid(True)

plt.show()
```

From the perspective of the JWST the flashes were simultaneous, but the probe saw the second flashlight flash first


```python
import numpy as np
import matplotlib.pyplot as plt


b = 2.898e-3


star_types = ['O5', 'B5', 'A5', 'F5', 'G5', 'K5', 'M5']
temperatures = [54000, 15200,8310, 6700, 5660, 4400, 3200]


wavelengths_max = [b / T for T in temperatures]

wavelengths_max_micrometers = [l * 1e6 for l in wavelengths_max]

plt.figure(figsize=(10, 6))
plt.plot(star_types, wavelengths_max_micrometers, marker='o', linestyle='-', color='b', label='Wavelength of Max Intensity')

plt.xlabel('Star Types')
plt.ylabel('Wavelength of Max Intensity (μm)')
plt.title("Wavelength of Maximum Intensity for Different Star Types")
plt.grid(True)
plt.legend()


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


h = 4.135667696e-15
Phi_Al = 4.2
Phi_Au = 5.1

v = np.linspace(0, 5e15, 1000)


v_cutoff_Al = Phi_Al / h
v_cutoff_Au = Phi_Au / h


K_max_Al = h * (v - v_cutoff_Al)
K_max_Au = h * (v - v_cutoff_Au)


plt.figure(figsize=(10, 6))
plt.plot(v, K_max_Al, label='Aluminum', color='blue')
plt.plot(v, K_max_Au, label='Gold', color='red')


plt.axvline(x=v_cutoff_Al, color='blue', linestyle='--', label='Aluminum Cutoff Frequency')
plt.axvline(x=v_cutoff_Au, color='red', linestyle='--', label='Gold Cutoff Frequency')


plt.xlabel('Frequency (Hz)')
plt.ylabel('Max Kinetic Energy (eV)')
plt.title('Maximum Photoelectron Kinetic Energy vs Frequency')
plt.legend()
plt.grid(True)

plt.show()
```
